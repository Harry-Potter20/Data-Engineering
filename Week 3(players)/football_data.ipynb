```python
import requests 
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd
```

```python
# Lets create the url variable to load the relevant player data 
url = "https://www.worldfootball.net/players_list/eng-premier-league-2024-2025/nach-name/1/"
ua = UserAgent()
userAgent = ua.random
headers = {'User-Agent': userAgent}
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
Table = soup.find('table', class_="standard_tabelle")
Table
```

```python
headers = []
for i in Table.find_all('th'):
    title = i.text.strip()
    headers.append(title)

premier_league_players  = pd.DataFrame(columns = headers)

for j in Table.find_all('tr')[1:]:
    row_data = j.find_all('td')
    row = [i.text for i in row_data]
    length = len(premier_league_players)
    premier_league_players.loc[length] = row

premier_league_players
```

```python
# Lets clean the dataframe 
premier_league_players = premier_league_players.rename(columns={'born': 'Date_of_Birth', 'Height': 'Height(cm)'})
premier_league_players

# Change datatype of 'Height'
premier_league_players['Height(cm)'] = premier_league_players['Height(cm)'].astype('str')
premier_league_players['Height(cm)'] = premier_league_players['Height(cm)'].str.replace(' cm', '')
premier_league_players['Height(cm)'] = premier_league_players['Height(cm)'].str.replace('???', '1')
premier_league_players['Height(cm)'] = premier_league_players['Height(cm)'].astype('int')
premier_league_players['Height(cm)'] = premier_league_players['Height(cm)']

# Lets work on the height more
mean_height = premier_league_players[premier_league_players['Height(cm)'] != 1]['Height(cm)'].mean()
premier_league_players['Height(cm)'] = premier_league_players['Height(cm)'].replace(1, mean_height)
premier_league_players['Height(cm)'] = premier_league_players['Height(cm)'].astype('int')
premier_league_players
premier_league_players.info()



premier_league_players
```
