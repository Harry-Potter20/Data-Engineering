```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd
```

```python
url = 'https://www.jumia.com.ng/ovens-toasters/'
ua = UserAgent()
userAgent = ua.random
headers = {'User-Agent': userAgent}
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
ovens = soup.find_all('div', class_="info")
items = []
curr_prices = []
old_prices = []

for oven in ovens:
    if oven.find('h2', class_="name") is not None:
        item = oven.find('h2', class_="name")
        items.append(item.text)
    else:
        items.append(None)
print(items)

for oven in ovens:
    if oven.find('span', class_="curr") is not None:
        price = oven.find('span', class_="curr")
        curr_prices.append(price.text)
    else:
        items.append(None)

for oven in ovens:
    if oven.find('span', class_="old") is not None:
        old = oven.find('span', class_="old")
        old_prices.append(old.text)
    else:
        old_prices.append(None)

print(ovens)
print(curr_prices)
print(old_prices)

ovens_df = pd.DataFrame({'item':items, 'new_price':curr_prices, 'old_price':old_prices})

ovens_df
```
